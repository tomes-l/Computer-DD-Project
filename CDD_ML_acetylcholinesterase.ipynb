As the treshold between active and inactive molecules based on the IC50 is arbitrary, and only specific to the context, we are going to look at the values in our database to determine 3 different classes (active, inactive, intermediate).

```python
df_bioactivity_data.dtypes
```

```python
df_bioactivity_data['standard_value'] = df_bioactivity_data['standard_value'].astype(float)    
df_bioactivity_data['standard_value'].describe()
```

It is clear that all molecules, for which IC50 can be determined, are (in principle) "active". But in fact the higher the value, the higher the dose needed to achieve the inhibitory effect and the higher the chance of "off-target" drug binding events, and hence, potential toxicity.

### For that sake, and in keeping in mind that we don't need to satisfy biological treshold, we'll consider based on the distribution :
- compounds having values of less than 2000 nM as active
- compounds having values greater than 15,000 nM as inactive
- compounds having values in between 2,000 and 15,000 nM as intermediate

```python
bioactivity_threshold = []

for value in df_bioactivity_data['standard_value']:
  if value >= 15000:
    bioactivity_threshold.append("inactive")
  elif value <= 2000:
    bioactivity_threshold.append("active")
  else:
    bioactivity_threshold.append("intermediate")
```

```python
bioactivity_class = pd.Series(bioactivity_threshold, name='class')
df_bioactivity = pd.concat([df_bioactivity_data, bioactivity_class], axis=1)

df_bioactivity.dropna(inplace=True)
df_bioactivity
```

### Convert IC50 to pIC50

In the following, we will :
- Take the IC50 values from the standard_value column and converts it from nM to M by multiplying the value by 10
- Take the molar value and apply -log10
- Delete the standard_value column and create a new pIC50 column

**But first let's have a look at our values**

```python
df_combinedf['standard_value'].describe()
```

We can note that :

```python
-np.log10((10**-9) * (10**9))
```

```python
-np.log10((10**-9)* (10**-3))
```

In that way, we are sure to have only positive values in the distribution from a range to **0 till 12** 

```python
# Inspired by : https://github.com/chaninlab/estrogen-receptor-alpha-qsar/blob/master/02_ER_alpha_RO5.ipynb

def pIC50(df0):
    pIC50 = []

    for i in df0['standard_value']:
        molar = i*(10**-9) # Converts nM to M
        pIC50.append(-np.log10(molar))

    df0['pIC50'] = pIC50
    df0 = df0.drop(columns=['standard_value'], axis=1)
        
    return df0
     
```

```python
df_final = pIC50(df_combinedf)
df_final
```

## Calculate morgan fingerprint descriptors from RDKit

### Importing libraries

```python
from rdkit.Chem import rdFingerprintGenerator
```

We know that if we want to be able to improve the accuracy of the fp representation (and so of the prediction), we have to increase its size. But if we do that, the computational cost can become too big.

### That's why we set the fp size at **500 bits**, and we will use later a method to remove low variance features

```python
len(smiles)
```

```python
# Initialize all keys with empty lists
fingerprint = {f"FP{i+1}": [] for i in range(len(smiles))}
n = 1

for idx, elem in enumerate(smiles):
    morganfpgen = rdFingerprintGenerator.GetMorganGenerator(radius=2, fpSize=500)
    mol = Chem.MolFromSmiles(elem)
    fp = morganfpgen.GetFingerprint(mol)

    # Convert the fingerprint to a list
    fp_list = list(fp) 
    
    for idx, bit in enumerate(fp_list):
        # Append each bit value to its corresponding key's list
        fingerprint[f"FP{n}"].append(bit)
    n += 1

df_fingerprint = pd.DataFrame.from_dict(fingerprint)

df_fingerprint
```

## Remove low variance features

### Importing libraries

```python
!pip install scikit-learn
```

```python
import sklearn
from sklearn.feature_selection import VarianceThreshold
```

### We inspect the variance of the fingerprints to choose the right threshold

```python
variance = df_fingerprint.var()

plt.figure(figsize=(10, 6))
plt.hist(variance, bins=30, edgecolor='k', alpha=0.7)
plt.xlabel('Variance', fontsize=16, fontweight='bold')
plt.ylabel('Frequency', fontsize=16, fontweight='bold')
plt.show()
```

We see that the majority of the variance is around **0.08**. Also, a good threshold could be around **0.04** to eliminate low variance features but keep the majority of the usefull data.

```python
selector = VarianceThreshold(threshold=(0.04 * (1 - 0.04)))
selection = selector.fit_transform(df_fingerprint)

df_fp = pd.DataFrame(selection)

df_fp
```
